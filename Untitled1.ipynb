```python

import json
import matplotlib.pyplot as plt
import re

# Define all τDT values and matching file paths
result_files = {
    0.1: "../runs/exp_dtslru_2024/20230325_Region5_0_0.1_366.475GB_WR35.599MBS/dtslru_tau_0.1/rejectx-1_lru_366.475GB/full_0_0.1_cache_perf.txt.stats",
    0.5: "../runs/exp_dtslru_2024/20230325_Region5_0_0.1_366.475GB_WR35.599MBS/dtslru_tau_0.5/rejectx-1_lru_366.475GB/full_0_0.1_cache_perf.txt.stats",
    1.0: "../runs/exp_dtslru_2024/20230325_Region5_0_0.1_366.475GB_WR35.599MBS/dtslru_tau_1.0/rejectx-1_lru_366.475GB/full_0_0.1_cache_perf.txt.stats",
    2.0: "../runs/exp_dtslru_2024/20230325_Region5_0_0.1_366.475GB_WR35.599MBS/dtslru_tau_2.0/rejectx-1_lru_366.475GB/full_0_0.1_cache_perf.txt.stats",
    5.0: "../runs/exp_dtslru_2024/20230325_Region5_0_0.1_366.475GB_WR35.599MBS/dtslru_tau_5.0/rejectx-1_lru_366.475GB/full_0_0.1_cache_perf.txt.stats"
}

taus = sorted(result_files.keys())
hit_rates = []

for tau in taus:
    file_path = result_files[tau]
    try:
        with open(file_path, 'r') as f:
            content = f.read()
        
        # Try different approaches to find hit rate
        hit_rate = None
        
        # Approach 1: Look for JSON structure
        try:
            data = json.loads(content)
            # Try common JSON structures for cache performance
            if "hit_rate" in data:
                hit_rate = data["hit_rate"]
            elif "hit_rate_percent" in data:
                hit_rate = data["hit_rate_percent"]
            elif "cache_stats" in data and "hit_rate" in data["cache_stats"]:
                hit_rate = data["cache_stats"]["hit_rate"]
            elif "overall" in data and "hit_rate" in data["overall"]:
                hit_rate = data["overall"]["hit_rate"]
        except:
            # If JSON parsing fails, try text parsing
            pass
        
        # Approach 2: Text parsing with regex
        if hit_rate is None:
            # Look for hit rate patterns in text
            hit_rate_match = re.search(r'Hit Rate:?\s*([0-9.]+)', content, re.IGNORECASE)
            if hit_rate_match:
                hit_rate = float(hit_rate_match.group(1))
            
            # Look for percentage patterns
            percent_match = re.search(r'([0-9.]+)%', content)
            if percent_match and hit_rate is None:
                hit_rate = float(percent_match.group(1))
            
            # Look for flash hit rate specifically
            flash_hit_match = re.search(r'Flash Hit Rate:?\s*([0-9.]+)', content, re.IGNORECASE)
            if flash_hit_match:
                hit_rate = float(flash_hit_match.group(1))
        
        if hit_rate is not None:
            hit_rates.append(hit_rate)
            print(f"τDT={tau}: Hit Rate = {hit_rate}%")
        else:
            print(f"τDT={tau}: Could not find hit rate data")
            hit_rates.append(None)
            
    except Exception as e:
        print(f"Error reading τDT={tau}: {e}")
        hit_rates.append(None)

# Filter out missing values
taus_filtered = [t for t, h in zip(taus, hit_rates) if h is not None]
hit_rates_filtered = [h for h in hit_rates if h is not None]

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(taus_filtered, hit_rates_filtered, marker='o', linestyle='-', color='green', linewidth=2, markersize=8)
plt.title("Figure 2: Hit Rate vs τDT (DT-SLRU)", fontsize=14)
plt.xlabel("τDT (Promotion Threshold)", fontsize=12)
plt.ylabel("Hit Rate (%)", fontsize=12)
plt.grid(True, alpha=0.3)
plt.xscale('log')  # Logarithmic scale for better visualization
plt.tight_layout()
plt.savefig('figures/fig2_hit_rate_vs_tau_dt.png', dpi=300, bbox_inches='tight')
plt.show()

# Print summary
print("\n=== Summary for Figure 2 ===")
for tau, hr in zip(taus_filtered, hit_rates_filtered):
    print(f"τDT = {tau}: Hit Rate = {hr:.2f}%")
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np

# Correct result file paths for τDT values
result_files = {
    0.1: "../runs/exp_dtslru_2024/20230325_Region5_0_0.1_366.475GB_WR35.599MBS/dtslru_tau_0.1/rejectx-1_lru_366.475GB/full_0_0.1_cache_perf.txt.stats",
    0.5: "../runs/exp_dtslru_2024/20230325_Region5_0_0.1_366.475GB_WR35.599MBS/dtslru_tau_0.5/rejectx-1_lru_366.475GB/full_0_0.1_cache_perf.txt.stats",
    1.0: "../runs/exp_dtslru_2024/20230325_Region5_0_0.1_366.475GB_WR35.599MBS/dtslru_tau_1.0/rejectx-1_lru_366.475GB/full_0_0.1_cache_perf.txt.stats",
    2.0: "../runs/exp_dtslru_2024/20230325_Region5_0_0.1_366.475GB_WR35.599MBS/dtslru_tau_2.0/rejectx-1_lru_366.475GB/full_0_0.1_cache_perf.txt.stats",
    5.0: "../runs/exp_dtslru_2024/20230325_Region5_0_0.1_366.475GB_WR35.599MBS/dtslru_tau_5.0/rejectx-1_lru_366.475GB/full_0_0.1_cache_perf.txt.stats"
}

# τDT values
taus = [0.1, 0.5, 1.0, 2.0, 5.0]

# Metrics to plot
metric_keys = {
    "puts_ios": "puts_ios/ns/1476116018_stats",
    "puts_chunks": "puts_chunks/ns/1476116018_stats",
    "service_time_nocache": "service_time_nocache/ns/1476116012_stats"
}

# Plot each metric
for label, key in metric_keys.items():
    metric_values = []
    for tau in taus:
        try:
            with open(result_files[tau]) as f:
                data = json.load(f)
                values = data.get(key, [])
                if not values:
                    raise ValueError(f"Missing data for key '{key}'")
                avg = np.mean(values)
                metric_values.append(avg)
        except Exception as e:
            print(f"[τDT={tau}] Error: {e}")
            metric_values.append(None)

    # Filter missing data
    taus_filtered = [t for t, v in zip(taus, metric_values) if v is not None]
    values_filtered = [v for v in metric_values if v is not None]

    # Plot
    plt.figure(figsize=(8, 5))
    plt.plot(taus_filtered, values_filtered, marker='o', linestyle='-')
    plt.title(f"Figure: Avg. {label} vs τDT")
    plt.xlabel("τDT")
    plt.ylabel(f"Average of {label}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
